## Loading all the libraries needed

```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series, read_csv
import matplotlib.pyplot as plt
import matplotlib 

# Enable inline plotting
%matplotlib inline
```

## Loading File

```python
location = r'high-protien-food.csv'
df = pd.read_csv(location)
```

### Cleaning out empty rows and colums

```python
df = df.dropna()
df = df.dropna(axis=1)
df
```

## Top 5 Lowest Protien

```python
df.sort(["Protein"], axis=0, ascending=True, inplace=True)
df.head(5)
```

## Food with Highest Protien

```python
df.sort(["Protein"], axis=0, ascending=False, inplace=True)
df.head(5)
```

## Food With Highest Fat

```python
df.sort(["Fat g"], axis=0, ascending=False, inplace=True)
df.head(5)
```

## Food With Lowest Fat

```python
df.sort(["Fat g"], axis=0, ascending=True, inplace=True)
df.head(5)
```

## Food With Highest Carb

```python
df.sort(["Carbs g"], axis=0, ascending=False, inplace=True)
df.head(5)
```

## Food With Lowest Carb

```python
df.sort(["Carbs g"], axis=0, ascending=True, inplace=True)
df.head(5)
```
